```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import * 
```

```python
x = np.array([np.sqrt(1/252), np.sqrt(5/252), np.sqrt(21/252), np.sqrt(63/252), np.sqrt(126/252), np.sqrt(252/252)])
y = np.array([1.461453224, 2.954049141, 5.580549554, 9.557047518, 14.04933231, 20.77227112])
m, b = np. polyfit(x, y, 1)
```

```python
plt.plot(x,y,'o', color='black')
plt.plot(x, m*x+b)
plt.xlabel('Square-root of Time')
plt.ylabel('Standard Deviation') 
plt.title('The Visualization graph of the standard deviation of EWJ returns as a function of time')
```
